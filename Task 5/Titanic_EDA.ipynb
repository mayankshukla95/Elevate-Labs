Question 2. Use sns.pairplot(), sns.heatmap() for visualization.

Pairplot

```python
import seaborn as sns
sns.pairplot(df,hue='Survived')
```

Observation :-  


1.   Correlation Between Higher Fares and Survival: Passengers who paid higher fares generally have a higher survival rate. This is noticeable as the data points for survivors are more concentrated in the upper ranges of the Fare axis.
2.   Survival and Passenger Class: Passengers in higher classes (Pclass 1) show a notably higher survival rate compared to those in lower classes (Pclass 3). This is evident in the clustering of survivor data points in Pclass 1.



Heatmap

```python
import matplotlib.pyplot as plt
num_df=df.select_dtypes(include=['float64','int64'])
correlation_matrix = num_df.corr()
sns.heatmap(correlation_matrix, annot = True, cmap = 'coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

Observation :-  


1.   Pclass and Fare Relationship: There's a strong negative correlation (-0.55) between 'Pclass' and 'Fare'. This suggests that passengers in higher ticket classes (e.g., 1st class) paid significantly higher fares compared to passengers in lower classes (e.g., 3rd class).
2.   Fare and Survival Relationship: A moderate positive correlation (0.26) exists between 'Fare' and 'Survived'. This indicates that passengers who paid higher fares had a better chance of survival.

Question 4. Plot histograms, boxplots, scatterplots

Histogram

```python
sns.histplot(data=df, x='Age', bins=30, kde=True, color='blue')
plt.title('Age Distribustion')
plt.show()
```

Observation :-  


1.   Age Group Concentration: The age distribution peaks around the 20-30 year range, indicating that the majority of individuals fall within this age bracket. This represents the densest population segment.
2.   Decrease in Older Age Groups: As age increases beyond 30 years, there is a noticeable decline in the count of individuals. Very few individuals are observed in the age group above 70 years.

Boxplot

```python
sns.boxplot(data=df, x='Survived', y='Age')
plt.title('Age Distribustion By Survival')
plt.show()
```

Observation :-  


1.   Median Age Difference by Survival: The median age of survivors (represented by 1) is slightly lower than the median age of non-survivors (represented by 0). This indicates that younger individuals might have had a higher survival rate.
2.   Outlier Analysis: There are more outliers in the age distribution of non-survivors compared to survivors. Non-survivors show age outliers extending into higher ranges, while the oldest survivor's age is around 80 years.

Scatter Plot

```python
sns.scatterplot(data=df, x='Age', y='Fare', hue='Survived', style='Pclass')
plt.title('Fare vs. Age by Survival')
plt.show()
```

Observation :-  


1.   Higher Fare Indicates Higher Survival Rates: Passengers who paid fares above 100 were more likely to survive (orange points), as seen in their prominent clustering in this range. This trend persists across different age groups.
2.   Majority Paid Lower Fares: Most passengers, regardless of survival status, are concentrated in the lower fare range (below 100). This suggests that lower-paying fares were more common, with varying survival outcomes across different passenger classes.

# Summary : -   
Passengers in higher classes, who paid higher fares, were more likely to survive. Younger individuals also showed higher survival rates. Most passengers were concentrated in the 20-30 age range and paid lower fares, but survival was skewed towards those with higher fares and better class. Gender and age played roles in survival outcomes, with variability seen in the data.
